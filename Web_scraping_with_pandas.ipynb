# Web scraping project with pandas

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import numpy as np
```

```python
# Copying url of table we want to analyze
url="https://www.basketball-reference.com/leagues/NBA_1998_per_game.html"
```

```python
# Using pd.read_html toread HTML webpage
df_url = pd.read_html(url, header = 0)
df_url
```

We can see that we have 100 null values in "3P%" column so let's investigate this.

```python
a["3P"].value_counts()
```

We can see that the reason why we have null values in "3P%" column is that are players that haven't score 3-points. We will replace null values with 0.

```python
df3=df2.copy()
```

```python
# Replacing values
df3["3P%"]=df2["3P%"].replace(np.nan,"0")
```

```python
df3.isnull().sum()
```

```python
df3["Tm"].unique()
```

```python
#Distribution plot of points by players
sns.distplot(df2["PTS"], 
             kde=False,
             hist_kws=dict(edgecolor="black", linewidth=2))
```

We have over 100 players that score between 0 and 5.

```python
df2["Age"]=df2["Age"].astype(int)
```

```python
#Defining function
def age(row):
    if row["Age"]<=20:
        return "Rookie"
    if(20 < row["Age"]<=25):
        return "Junior"
    if(25 < row["Age"]<=30):
        return "Senior"
    else:
        return "Experienced"
```

```python
#Applying function to  new column
df2["Experience_player"]=df2.apply(age,axis=1)
df2.head()
```

```python
#Countplot
sns.countplot(x="Experience_player",hue="Pos",data=df2)
```

We can see that rookies don't play a few position. Players in best years play PF the most.

```python
# Pie chart
plt.pie(x=df2['Experience_player'].value_counts(), labels=df2["Experience_player"].value_counts().index, autopct='%1.1f%%')
plt.show()
```

```python
df3=df2.copy()
```

```python
#Converting to float
df3["PTS"]=df2["PTS"].astype(float)
```

Let's compare Chicago Bulls with LA Lakers,Utah Jazz and Detroit Pistons based on 2-point field goal percentage,3-point field goal percentage and free throw percentage. 

```python
chi_la_det=df5[(df5["Tm"]=="CHI") |(df5["Tm"]=="UTA") | (df5["Tm"]=="DET")]
team=chi_la_det["Tm"].unique()
team
```

```python
xpos=np.arange(len(team))
```

```python
team_nba=chi_la_det.copy()
```

```python
team_nba["ORB"]=chi_la_det["ORB"].astype(float)
team_nba["DRB"]=chi_la_det["DRB"].astype(float)
team_nba["TRB"]=chi_la_det["TRB"].astype(float)
```

```python
chi=team_nba.copy()
```

```python
team_chi=chi[chi["Tm"]=="CHI"]
```

```python
offensive_rebounds_chi=team_chi.ORB.mean()
defensive_rebounds_chi=team_chi.DRB.mean()
total_rebounds_chi=team_chi.TRB.mean()
```

```python
utah=team_nba.copy()
```

```python
team_utah=utah[utah["Tm"]=="UTA"]
```

```python
offensive_rebounds_ut=team_utah.ORB.mean()
defensive_rebounds_ut=team_utah.DRB.mean()
total_rebounds_ut=team_utah.TRB.mean()
```

```python
det=team_nba.copy()
```

```python
team_det=det[det["Tm"]=="DET"]
```

```python
offensive_rebounds_det=team_det.ORB.mean()
defensive_rebounds_det=team_det.DRB.mean()
total_rebounds_det=team_det.TRB.mean()
```

```python
plt.figure(figsize=(14,8))
plt.xticks(xpos,team)
plt.title("Comparison Chicago Bulls with Utah Jazz and Detroit Pistons by rebounds")
plt.bar(xpos,[offensive_rebounds_ut,offensive_rebounds_chi,offensive_rebounds_det],width=0.2, label="Offensive rebounds per game")
plt.bar(xpos-0.2,[defensive_rebounds_ut,defensive_rebounds_chi,defensive_rebounds_det],width=0.2, label="Defensive rebounds per game")
plt.bar(xpos+0.2,[total_rebounds_ut,total_rebounds_chi,total_rebounds_det],width=0.2, label="Total rebounds per game")
plt.legend(loc="best",framealpha=0.2)
```

#### Finding best rookie,junior,senior and experienced player

```python
df6=df5.copy()
```

```python
df6["AST"]=df5["AST"].astype(float)
```

```python
#Creating column based on points,block,steal and assist per game
df6["Rating"]=df6["PTS"]+df6["BLK"]+df6["STL"]+df6["AST"]
df6.head()
```

The best player will be selected by higher "Rating" score.

```python
#Cretaing datasets based on player experience
rookie=df6[df6["Experience_player"]=="Rookie"]
junior=df6[df6["Experience_player"]=="Junior"]
senior=df6[df6["Experience_player"]=="Senior"]
experienced=df6[df6["Experience_player"]=="Experienced"]
```

```python
best_rookie=rookie.sort_values("Rating",ascending=False)
best_rookie.head()
```

```python
best_junior=junior.sort_values("Rating",ascending=False)
best_junior.head()
```

```python
best_senior=senior.sort_values("Rating",ascending=False)
best_senior.head()
```

```python
best_experienced=experienced.sort_values("Rating",ascending=False)
best_experienced.head()
```
